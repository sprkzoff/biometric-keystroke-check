# Define biometric checking by using MSE

```python
from sklearn.metrics import mean_squared_error

def check_biometric(attemp) :
    my_biometric = data.iloc[:,1]
    if mean_squared_error(my_biometric,attemp) <= 0.1 :
        print('Pass!! welcome..')
    else :
        print('Get out hacker!!')

hacker = data.iloc[:,2]
```

```python
me = data.iloc[:,3]
check_biometric(me)
```
